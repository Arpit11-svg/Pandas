```python
# TASK: create a function when we enter name of player then that can return the team which against that player take most runs
def run_scored(batsman_name):
    vk=delivery[delivery['batsman']==batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
```

```python
run_scored("MS Dhoni")
```

```python
# isin() function

# TASK: find the most destructive death over batsman in the history of IPL
# Strike rate=(Number of runs/Number of balls)/100=> Or how many runs he take in 100 balls
# Min batsman 200 balls in over 16-20=> he have atleast played 200 balls
```

```python
death_over=delivery[delivery['over']>15] #gives player who have played more than 15 overs or 15*6 balls
```

```python
all_batsman=death_over.groupby('batsman')['batsman_runs'].count()
```

```python
x=all_batsman>200 #but we need player name , these are ball or number of rows 
batsman_list=all_batsman[x].index.tolist() #gives name of player as index => we get all player which played more than 15 over
```

```python
# Now we have all player which played more than 15 over
# now we have to take these player only from delivery then use isin()
delivery[delivery['batsman'].isin(batsman_list)] # now we get only those data value which played by these player
```

```python
final=delivery[delivery['batsman'].isin(batsman_list)]
```

```python
runs=final.groupby('batsman')['batsman_runs'].sum() #total run of these player
balls=final.groupby('batsman')['batsman_runs'].count() #total run of these player
```

```python
strike_rate=(runs/balls)*100
```

```python
strike_rate.sort_values(ascending=False).head()
```
